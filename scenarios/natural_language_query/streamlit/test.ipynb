```python
import pyodbc
import pandas as pd
import numpy as np
import sqlite3
def execute_sql_query(query):
    db_path ='../data/northwind.db'
    conn = sqlite3.connect(db_path)

    cursor = conn.cursor()
    try:
        cursor.execute(query) 
        data =cursor.fetchall()
    except Exception as e:
        return str(e)

    cols = []
    for i,_ in enumerate(cursor.description):
        cols.append(cursor.description[i][0])
    if len(cols)>0:
        result = pd.DataFrame(np.array(data), columns = cols).head(20) #limit to 20
    else:
        result = pd.DataFrame()

    return result
```

```python
execute_sql_query("select * from employees limit 10")
```
